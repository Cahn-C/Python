# <b>The Weather Dataset: Reading DataFrames from Files</b>

The Weather Dataset is a time-series data set with per-hour information about the weather conditions at a particular location. It records Temperature, Dew Point Temperature, Relative Humidity, Wind Speed, Visibility, Pressure and Conditions.

```python
import pandas as pd
```

```python
# Read the data into a data frame

weather_df = pd.read_csv("/content/sample_data/weather_2012.csv")

print(weather_df.shape)
print(weather_df.index)
```

```python
# Get the first five rows

weather_df.head()
```

```python
# Get the first five rows from a single column

weather_df['Date/Time']
```

```python
weather_df['Date/Time'] = pd.to_datetime(weather_df['Date/Time'])
```

```python
print(weather_df['Date/Time'])
```

## <b>`.info()`</b>

Provides a summary of a DataFrame rows, columns data types of columns (if automatically detected) and the memory usage

For detailed summaries of the DataFrame, you can pass optional arguments `verbose=True` and `null_counts=True` to the .info() method to output information for all of the columns

```python
weather_df.info()
```

## <b>`.index`</b>

This attribute provides the index of the dataframe

Indexing identifies data using known indicators that allows intuitive getting and setting of subnets of the data set.

```python
weather_df.index
```

## <b>`.unique()`</b>

This method, which belongs to the `Series` object, can be useful when trying to identify unique values in a column.

- Uniques are returned in order of appearance.
- It is significantly faster than numpy.unique and includes N/A values.

```python
weather_df['Weather']
```

```python
weather_df['Weather'].unique()
```

```python
weather_df['Wind Spd (km/h)']
```

```python
weather_df['Wind Spd (km/h)'].unique()
```

## <b>`.nunique()`</b>

This method belongs to the `Series` object and can be useful when trying to identify the number of unique values in a column.

- Excludes NA values by default
- Always returns an integer value

```python
weather_df['Weather'].nunique()
```

## <b>`.value_counts`</b>

This method, belongs to the `Series` object, and can be useful when trying to identify unique values and their counts in a column

- The resulting object will be in descing order so that the first element is the most frequently-occurring element.
- Excludes NA values by default

```python
weather_df['Weather'].value_counts()
```

```python
# Get a single value

weather_df['Weather'].value_counts()[3]
```
